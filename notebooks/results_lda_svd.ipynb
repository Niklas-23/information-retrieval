# Plot Embeddings

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
```

```python
embeddings_answers= np.load("../arqmath_dataset/model_embeddings/lda_embedding_answers.npy")
embeddings_topics = np.load("../arqmath_dataset/model_embeddings/lda_topics_embedding.npy")
```

```python
pca = PCA(n_components=2)
answers = pd.DataFrame(pca.fit_transform(embeddings_answers))
topics = pd.DataFrame(pca.fit_transform(embeddings_topics))
```

```python
plt.scatter(answers[0], answers[1], label="Documents", c="lightgreen")
plt.scatter(topics[0], topics[1], label="Topics", c="blue")
plt.title("LDA embeddings")
plt.legend()
plt.savefig("lda_embedding.png", dpi=300)
plt.show()
```
